```python
import json
import random

sample_size = 1000
input_path = "watch-history.json"
output_path = "yt_sample.json"

with open(input_path, "r", encoding="utf-8") as f:
    records = json.load(f)

k = min(sample_size, len(records))
sampled_records = random.sample(records, k=k)

with open(output_path, "w", encoding="utf-8") as f:
    json.dump(sampled_records, f, ensure_ascii=False, indent=2)

print(f"Saved {k} random records to {output_path}")
```

```python
# load the sample and check it
with open(output_path, "r", encoding="utf-8") as f:
    sample_data = json.load(f)
sample_df = pd.DataFrame(sample_data)
sample_df.head()
```
